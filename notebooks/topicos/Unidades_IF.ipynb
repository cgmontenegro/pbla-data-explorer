```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PDA_2020_Unidades_Rede_Federal_EPCT.csv', sep=';', encoding='iso-8859-1')
df['ANO'] = pd.to_datetime(df['ANO'], format='%Y').dt.year
df.sort_values("ANO")
display(df)
```

```python
df_count = df['ANO'].value_counts().reset_index(name='COUNT').sort_values(['index'], ascending=True)
display(df_count)
```

```python
df_count = df_count.set_index('index')
df.plot()
plt.show()
```
